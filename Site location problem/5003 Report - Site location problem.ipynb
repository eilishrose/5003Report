```python
import csv
import matplotlib
```

```python
weight1 = 20
weight2 = 30
weight3 = 50

weightedgeology = []
weightedtransport = []
weightedpopulation = []
```

```python
# Import geology data
```

```python
f = open('geology.csv', newline='' )
geology = []
reader = csv.reader(f,  delimiter=',', quoting=csv.QUOTE_NONNUMERIC)
for row in reader: 
    rowlist = []
    geology.append(rowlist)
    for value in row:
        rowlist.append(value)
f.close()
```

```python
# Map geology data
```

```python
matplotlib.pyplot.ylim(530,0)
matplotlib.pyplot.xlim(0,335)
matplotlib.pyplot.imshow(geology)
matplotlib.pyplot.show()
```

```python
# Import transport data
```

```python
f = open('transport.csv', newline='' )
transport = []
reader = csv.reader(f,  delimiter=',', quoting=csv.QUOTE_NONNUMERIC)
for row in reader: 
    rowlist = []
    transport.append(rowlist)
    for value in row:
        rowlist.append(value)
f.close()
```

```python
# Map transport data
```

```python
matplotlib.pyplot.ylim(530,0)
matplotlib.pyplot.xlim(0,335)
matplotlib.pyplot.imshow(transport)
```

```python
#Import population data
```

```python
f = open('population.csv', newline='' )
population = []
reader = csv.reader(f,  delimiter=',', quoting=csv.QUOTE_NONNUMERIC)
for row in reader: 
    rowlist = []
    population.append(rowlist)
    for value in row:
        rowlist.append(value)
f.close()
```

```python
#Map population data
```

```python
matplotlib.pyplot.ylim(530,0)
matplotlib.pyplot.xlim(0,335)
matplotlib.pyplot.imshow(population)
```
